```python
import pandas as pd #Importing of the libraries we will use
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Averaged.csv') #Reading the Datafame
```

```python
df.shape
```

```python
df.head() #First 5 parts of data frame
```

```python
df.drop(['TEAM'], axis = 1) #Dropping the teams column
```

```python
df.tail() # Last 5 parts of our dataframe
```

```python
df.info() #Checking to see what type the value is
```

```python
df.describe() #Checking to see the mean, standard deviation, and the quartiles.
```

```python
df.isnull() #Checking to see if there is any null values, to clean our data
```

```python
df.isnull().sum() #There is no null value present
```

```python
df.dropna(inplace = True)
```

```python
target_column=['WINP'] #Our dependent variable
independent_variables=['GP','W','L','WINP','MIN','PTS','FGM','FGA','FGP','3PM','3PA',
            '3PP','FTM','FTA','FTP','OREB','DREB','REB','AST','TOV','STL','BLK','BLKA','PF','PFD','POS/NEG'] #Our independent variables

x=df[independent_variables].values
y=df[target_column].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20) #The Training and Test Split
```

```python
x_train.shape #The Size of our x_train
```

```python
y_train.shape #The Size of our y_train
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor() # Initializing our model
```

```python
rf.fit(x_train, y_train) #Fitting our training sets into our model for manipulation
```

```python
y_pred = rf.predict(x_test) #Getting the predicted value that we get from our model
```

```python
y_pred # Our predictions
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred) #Receiving the mean squared error
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
plt.scatter(y_test, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted') #Plotting our model
```

```python
pred_y_df=pd.DataFrame({'Actual Value': y_test.flatten(), 'Predicted Value': y_pred.flatten(), 'Difference': y_test.flatten()-y_pred.flatten()})
pred_y_df[0:20] #Checking the actual value and the predicted value of our model
```

```python
import math #These are our metrics

print("Metrics: ")

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R-Squared: ", r2)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error: ",mse)

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error: ",mae)

from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(y_test, y_pred)
print("Mean Absolute Percentage Error: ",mape)
print()
rmse = math.sqrt(mse)
print("The Root Mean Squared Error: ", rmse)

rmae = math.sqrt(mae)
print("The Root Mean Absolute Error: ", rmae)

rmape = math.sqrt(mape)
print("The Root Mean Absolute Percentage Error: ", rmape)
```
